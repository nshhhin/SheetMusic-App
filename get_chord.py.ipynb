# コードの図面を取得してくる

https://gakufu.gakki.me/code_g/ から頂戴してくる

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
```

```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
URL = "https://gakufu.gakki.me/code_g/"
```

```python
r = requests.get(URL)
soup = BeautifulSoup(r.content, "html.parser")
```

```python
codes = soup.find_all("img")
```

```python
for code in codes:
    src = code["src"]
    if 'https:' in src:
        print(src)
        splited = src.split("/")
        if len(splited) > 6:
            code_name = splited[6][0:-4]
            print(splited[6][0:-4])
            response = requests.get(src)
            image = response.content
            with open("chord_data/" + code_name + ".png", "wb") as file:
                file.write(image)
                time.sleep(1)
```
