```python
import os
import datetime
files = os.listdir('chord_data/')
```

```python
files.sort()
```

```python
def sort_mtime(rootdir):
    xs = []
    for root, dir, files in os.walk(rootdir):
        for f in files:
            path = os.path.join(root, f)
            xs.append((os.path.getmtime(path), path))

    for mtime, path in sorted(xs):
        name = os.path.basename(path)
        t = datetime.datetime.fromtimestamp(mtime)
        print(t, name)

def main():
    sort_mtime('chord_data/')

if __name__ == '__main__':
    main()
```

```python

import os
import glob
from datetime import datetime

_path = "chord_data/"

_files = glob.glob(_path)
_files.sort(
    cmp=lambda x, y: int(os.path.getctime(x) - os.path.getctime(y)),
    reverse=True)

for item in _files:
    _timestamp = os.path.getctime(item)
    print datetime.fromtimestamp(_timestamp).strftime("%Y/%m/%d %H:%M:%S")
```
